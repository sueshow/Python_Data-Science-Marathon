# 什麼是資料科學？

* 透過數據來解決問題的一門科學
* 根據資料、數據上的限制與環境需求為客戶選擇一個最合適的「計算模型」，解答「數理」問題，讓客戶得以根據該解法，轉譯出背後的商業價值，包含以下三個步驟：
  * 拆解問題
  * 組織問題
  * 解決問題

# 資料科學的名詞演進與用途

* 2012 年，資料科學家被《哈佛商業評論》稱為《二十一世紀最性感的職業後，資料科學逐漸成為一個時髦術語（英語：Buzzword)，但資料科學這個名詞和大數據、機器學習與人工智慧密不可分，但有些許不同
  * 大數據：資料收集
  * 資料科學：萃取商業價值/領域特徵
  * 機器學習/人工智慧：預測/分類

# 資料科學

## 領域範疇

* 資料科學的領域範疇，包含領域知識、程式設計及統計方法。如果你想在金融業擔任資料科學家，領域知識就是金融裡面的大小事與 know how，利用統計方法找出趨勢與因果關係，程式設計則是讓資料科學的效益能系統化實現。
* 4 種分析型態：資料科學所產生出的產品，包含描述型，診斷型、預測型與建議型，下圖根據價值和難度來做區分，可以看到描述行為基礎，而最終我們想得到建議型的產品，降低決策的風險與提高收益。
* 實作流程
  * 資料收集與定義：定義問題和目標→確定資料與分析任務→設計與收集資料(time ans resources)
  * 資料處理：探索性資料分析(EDA)/資料清理→特徵工程→建立預測模型/模型選擇評估準確度
  * 資料呈現：找出洞察力和含意→部署和表達進行決策應用(Domain Knowledge)

## 分析型態：探索性分析(EDA)

* EDA (Exploratory Data Analysis) 是由有統計界畢卡索之稱的 Tukey 所提出，主要概念是透過敘述性統計、統計繪圖、視覺化等快速簡易的方式，從各種面向先了解資料的狀況，以利後續分析。
* 任務：
  * 提早發現資料品質問題
  * 找出重要變數
* 步驟：
  * 確認資料型態(object、int、float)
  * 每個欄位的意義、分布
  * 檢測異常值和缺失值
  * 發掘特徵變數之間的關係
  * 測試基本的假設
  * 提取或找出重要特徵變數[特徵工程]
  * 初步選擇合適的模型
