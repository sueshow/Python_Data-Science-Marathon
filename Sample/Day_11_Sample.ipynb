# Day-11 Pandas 中的算術運算特性

* 教學目標：
  * 知道如何使用 DataFrame 中的運算
  * 了解 DataFrame 運算的特性
  * 知道 DataFrame 中統計與字串的操作

## 載入套件

```python
# 載入 NumPy, Pandas 套件
import numpy as np
import pandas as pd

# 檢查正確載入與版本
print(np)
print(np.__version__)
print(pd)
print(pd.__version__)
```

## 相同欄位的算術運算

* 在 DataFrame 的算術運算也有「對齊」的特性

```python
df1 = pd.DataFrame([[1, 2, 3]])
df2 = pd.DataFrame([[1, 1, 1]])

print(df1 + df2)
print(df1 - df2)
print(df1 * df2)
print(df1 / df2)
```

## 不同欄位的算術運算

* 在 DataFrame 的「對齊」的特性很嚴格，欄位對不上會產生錯誤的結果

```python
df1 = pd.DataFrame([[1, 2, 3]], columns=['a', 'b', 'c'])
df2 = pd.DataFrame([[1, 1, 1]], columns=['c', 'd', 'e'])

print(df1 + df2)
```

* 補充：NaN 是 Not a Number 的縮寫，在程式中泛指無法定義的數值

## DataFrame 也有廣播的特性

```python
df1 = pd.DataFrame([[1, 2, 3]])

print(df1 + 1)
```

## DataFrame 和 Array 有點不一樣

* 在 DataFrame 的「廣播」的特性也比較嚴格，只有支援常數的廣播

```python
df = pd.DataFrame([[1, 2, 3]])

print(df + 1)
print('=='*10)
print(df + pd.DataFrame([1]))
```

```python
a = np.array([[1, 2, 3]])

print(a + 1)
print('=='*10)
print(a + np.array([1]))
```

## 比較和邏輯運算

### DataFrame 也有遮罩運算

* DataFrame 也有沿用遮罩的特性，也可以用來做資料篩選

```python
print(df > 2)
print('=='*10)
print(df[df > 2])
```

```python
print(df['A'] > 2)
# a    False
# b     True
# Name: A, dtype: bool

print(df[df['A'] > 2])
#    A  B  C
# b  4  5  6
```

### DataFrame 中的排序

```python
df = pd.DataFrame({
    'col1': ['A', 'a', 'B', 'B', 'B', 'b'],
    'col2': [2, 1, 2, 3, 1, 8],
})

df.sort_values(by=['col1'])
```

```python
df.sort_values(by=['col1', 'col2'])
```

```python
df.sort_values(by='col2', ascending=False)
```

### DataFrame 的字串方法

* df.str.findall() ：對所有欄位的文字進行 re.findall()
* df.str.replace() ：對所有欄位的文字進行取代
* df.str.contains() ：對所有欄位的文字進行包含子字串的檢查
* df.str.count() ：對所有欄位的文字進行計數操作

## 匯入套件

```python
# 載入 NumPy, Pandas 套件
import numpy as np
import pandas as pd

# 檢查正確載入與版本
print(np)
print(np.__version__)
print(pd)
print(pd.__version__)
```

## 數值型資料

```python
score_df = pd.DataFrame([[1,50,80,70], 
              [2,60,45,50],
              [3,98,43,55],
              [4,70,69,89],
              [5,56,79,60],
              [6,60,68,55],
              [7,45,70,77],
              [8,55,77,76],
              [9,25,57,60],
              [10,88,40,43]],columns=['student_id','math_score','english_score','chinese_score'])
score_df = score_df.set_index('student_id')
score_df
```

### 平均數

* 每位學生的平均

```python
score_df.mean(axis=1)
```

* 指定欄位算平均

```python
#指定欄位算平均
score_df.math_score.mean()
```

#### 各科：平均數

```python
#全欄位算平均
score_df.mean()
```

### 3科總分數

```python
#學生3科總分數
score_df.sum(axis=1)
```

### 各科

#### 考試人數

```python
#本次各科考試人數
score_df.count()
```

#### 中位數

```python
#各科中位數分佈
score_df.median()
```

#### 百分位數

```python
#各科百分位數分佈(75%)
score_df.quantile(0.75)
```

#### 最大值 vs. 最小值

```python
#各科最大值
score_df.max()
```

```python
#各科最小值
score_df.min()
```

#### 標準差 vs. 變異數

```python
#各科標準差
score_df.std()
```

```python
#各科變異數
score_df.var()
```

#### 相關係數

```python
#各科之間的相關係數
score_df.corr()
```

### apply 用法

```python
#各科開根號乘以十
score_df.apply(lambda x : x**(0.5)*10)
```

```python
#各科加總apply
score_df.apply(sum,axis=1)
```
