## Series

```python
s = pd.Series([1, 2, 3])
print(s)
print(type(s))
```

### 常用屬性

```python
print(s.shape) # (3, )
print(s.size) # 3
print(s.dtype) # int64
```

### 範例

```python
s = pd.Series([1,2,3],  index=['Amy', 'Bob', 'Tom'])
print(s)
s
```

## DataFrame

```python
df = pd.DataFrame([1, 2, 3])
print(df)
print(type(df))
```

### 常用屬性

```python
print(df.shape) # (3, 1)
print(df.size) # 3
print(df.dtypes) 
```

```python
df = pd.DataFrame([1, 2, 3], index=['a', 'b', 'c'], columns=['No'])
print(df)
df
```

### 二維

```python
df = pd.DataFrame([[1, 2, 3], [4, 5, 6]], index=['a', 'b'], columns=['A', 'B', 'C'])
print(df)
df
```

### 範例

```python
df = pd.DataFrame([[1, 2, 3], [4, 5, 6]], index=['a', 'b'], columns=['A', 'B', 'C'])
print(df)
dfdf = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [18, 20],
})
print(df)
df
```

```python
df = pd.DataFrame([
  {'Name': 'Alice', 'Age': 18},
  {'Name': 'Bob', 'Age': 20}
])
print(df)
df
```

### DataFrame 是由 Series 组成的

```python
df = pd.DataFrame([[1, 2, 3], [4, 5, 6]], index=['a', 'b'], columns=['A', 'B', 'C'])
print(df['B'])
print(type(df['B']))
```

```python
print(df.shape)
print(df.size)
print(df.index)
print(df.columns)
print(df.values)
```

```python
print(df.head())
print(df.tail())
print(df.describe())
print(df.info())
```

### SQL

```python
import sqlite3
```

```python
boston_data = pd.read_excel('data.xls',sheet_name='boston',header=0 \
                            ,usecols=['TAX','PTRATIO','B','LSTAT'])
connection = sqlite3.connect('sql_db.sqlite')
boston_data.to_sql('boston', connection, if_exists='replace')
connection.commit()
connection.close()
```

```python
connection = sqlite3.connect('sql_db.sqlite')
boston_data_sql = pd.io.sql.read_sql("select * from boston",connection)
connection.close()
boston_data_sql
```
