$H_0$: $p_a$ = $p_b$<br>
$H_1$: $p_a$ $\neq$ $p_b$<br>

```python
A = [ 75, 30 ] 
B = [ 300, 300 ] 
result = statsmodels.stats.proportion.proportions_ztest( A, B, alternative='two-sided' )

print('Z-test: ', result[0])
print('P-value: ', result[1])
print('='*20)
print('på°æ–¼ 0.05 ç‚ºé¡¯è‘—æ‹’çµ• H0 å‡è¨­,å¯ä»¥èªªç”¢å“ä¸è‰¯ç‡ä¸åŒ')
```
